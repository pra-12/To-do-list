```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def display_tasks(self):
        if not self.tasks:
            print("Your to-do list is currently empty.")
        else:
            print("To-Do List:")
            for i, task in enumerate(self.tasks, start=1):
                status = "Completed" if task["completed"] else "Not Completed"
                print(f"{i}. {task['task']} ({status})")

    def add_task(self, task_name):
        self.tasks.append({"task": task_name, "completed": False})
        print(f"New task '{task_name}' added to your to-do list.")

    def mark_completed(self, task_number):
        if 1 <= task_number <= len(self.tasks):
            self.tasks[task_number - 1]["completed"] = True
            print(f"Task {task_number} marked as completed.")
        else:
            print("Invalid task number. Please enter a valid task number.")

    def remove_task(self, task_number):
        if 1 <= task_number <= len(self.tasks):
            removed_task = self.tasks.pop(task_number - 1)
            print(f"Task '{removed_task['task']}' has been removed from your to-do list.")
        else:
            print("Invalid task number. Please enter a valid task number.")


def main():
    to_do_list = ToDoList()

    while True:
        print("\nOptions:")
        print("1. Display your to-do list")
        print("2. Add a new task")
        print("3. Mark a task as completed")
        print("4. Remove a task from the list")
        print("5. Quit")

        choice = input("Enter the corresponding number for your choice: ")

        if choice == '1':
            to_do_list.display_tasks()
        elif choice == '2':
            task_name = input("Enter the new task: ")
            to_do_list.add_task(task_name)
        elif choice == '3':
            to_do_list.display_tasks()
            task_number = int(input("Enter the task number you completed: "))
            to_do_list.mark_completed(task_number)
        elif choice == '4':
            to_do_list.display_tasks()
            task_number = int(input("Enter the task number to remove: "))
            to_do_list.remove_task(task_number)
        elif choice == '5':
            print("Thank you for using the to-do list program. Have a great day!")
            break
        else:
            print("Invalid choice. Please enter a number corresponding to the options provided.")


if __name__ == "__main__":
    main()
```
